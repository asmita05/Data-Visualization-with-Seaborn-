# SEABORN

```python
import seaborn as sns
```

```python
%matplotlib inline
```

# DISTRIBUTION PLOT

```python
#visulallizing the distribution of dataset
#plotting univariate distribution
#tips is a buitin data set of seaborn and we save this dataframe with name tips
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#distribution of univariate(one variable) set of observation.
#you can pass only one column of dataframe in arguement.
#it allow us to visuallize distribution.
#we can see from graph that most of the bills lie between $10 to $20.
#if there is more datapoints in a given bin, then that bin will b high.
sns.distplot(tips['total_bill'],kde= False,bins=30)
```

```python
#pass rugplot() func or pass rug = True as a arguement in distplot to  see the density of observation instead of kde
#it appears as  a small  vertical tick at each observation to show the density
sns.distplot(tips['total_bill'], kde= False, bins=30, rug= True)
```

```python
#kde
sns.distplot(tips['total_bill'], hist=False, kde= True,  rug=True)
```

```python
#plotting bivariate distribution
#combine two diff distplot and it derive a relation between two variable
#most common way of visualizing bivariate distribution is scatterplot
#for this we use joint plotand scatterplot is within jointplot
#for ex-if we have to draw relation between total_bill and tips

sns.jointplot(x='total_bill', y='tip', data= tips)
```

```python
#if we chnage the kind  parameter kind : 'scatter', 'kde', 'reg', 'hex','resid'
#this kind parameter affects what going inside jointplot
sns.jointplot('total_bill','tip', data=tips, kind= 'hex')
```

```python
sns.jointplot('total_bill','tip', data= tips, kind ='kde')
```

```python
#pairplot() - it plots pairwise relationship in a dataset atleast for numerical columns
# it is going to do jointplot for every possible combination of a numerical columns in the dataset
#it draws scatter for joint relationship and histogram for univariate distribution. 
#for ex- tips-tips (histogram), tips-total_bill(scatter)
sns.pairplot(data= tips)
```

```python
#hue parameter used to pass  column name ie, categorical column not numerical or continuous
#for ex- sex column, there is 2 category, male or female
sns.pairplot(tips, hue= 'sex')
```

```python
#palette is used to cahnge the color of the parameters (husl,coolwarm)
sns.pairplot(tips, hue= 'sex',palette= 'husl')
```

```python
#rugplot() - it is used to plot datapoints in an array as sticks on the axis.
#it draws a vertical line(dashes) for every data points on the axis.
#pass single column in it.

sns.rugplot(tips['total_bill'])
```

```python
#sum of all the normal(gaussian) distribution represents kernel distribution estimation (kde). 
#normal distribution is built on the top of each dash in the rugplot.
#normal distribution for every datapoints is centered arond the dash.
#seaborn.kdeplot() - plot univariate and bivariate kernel density estimate.

sns.kdeplot(tips['total_bill'])
```

```python
#if shade= True, then it shade the area under the kde curve
sns.kdeplot(tips['total_bill'], shade= True,color= 'r')
```

```python
#plot a bivariate density estimate
sns.kdeplot(tips['total_bill'], tips['tip'], shade= True)
```

```python
#plot the density on the vertical axis.
sns.kdeplot(tips['tip'], vertical = True)
```

# CATEGORICAL PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

#### BAR PLOT

```python
# barplot() - it is used for the distribution of categorical column.
#here , mean of total_bill of (male & female) is shown.
# bydefault, mean is shown, but we can change it by estimator parameter, like if we want to draw std instead of mean.
#height of each rectangle represents the central tendency of numeric value (for ex - mean, std etc).
# x is  a categorical column and y is a numerical column.

sns.barplot(x='sex', y= 'total_bill', data= tips)
```

```python
# estimator- statistical function to estimate within each categorical bins.
import numpy as np
sns.barplot(x= 'sex', y= 'total_bill', data= tips, estimator= np.std)
```

#### COUNT PLOT

```python
#it is similar to bar plot except there is no y column because it just count the no of occurences in each categorical bins.

sns.countplot(x = 'sex', data= tips)

#it shows that there is more number of males than females in the dataset.
```

#### BOX PLOT

```python
#this plot shows the distribution of categorical data
#x= categorical column. y= numerical column
sns.boxplot(x= 'day', y='total_bill', data= tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

#### VIOLIN PLOT

```python
#IT ALSO SHOWS distibution of categorical data
#x= categorical column, y= numerical column

sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data= tips, hue= 'smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data= tips, hue='smoker', split= True)
```

```python
#show each observation with stick inside the violin
sns.violinplot(x='day', y='total_bill', data= tips, hue='smoker', split= True, inner='stick')
```

```python
#scale the density relative to the count across all bins
sns.violinplot(x='day', y='total_bill', data= tips, hue='smoker', split= True, scale='count', inner='stick', scale_hue= False)
```

#### STRIP PLOT

```python
#draw a scatterplot where on evariable is categorical.
sns.stripplot(x= 'day', y='total_bill', data= tips)
```

```python
#points are not clearly visible as they are overlapping, so add a parameter jitter=True
sns.stripplot(x= 'day', y='total_bill', data= tips, jitter= True)
```

```python
sns.stripplot(x= 'day', y='total_bill', data= tips, jitter= True, hue='sex')
```

```python
sns.stripplot(x= 'day', y='total_bill', data= tips, jitter= True, hue='sex', dodge= True)
```

```python
#draw stripplot on the top of box plot
pt= sns.boxplot(x='day', y='total_bill', data= tips, hue='sex', color='.3')
pt= sns.stripplot(x='day',y='total_bill', data= tips, jitter= True, hue='sex', dodge= True)
```

```python
#draw strip plot on the top of violin plot
pt= sns.violinplot(x='day', y='total_bill', data= tips, hue='sex', color='.3',split=True)
pt= sns.stripplot(x='day',y='total_bill', data= tips, jitter= True, hue='sex', dodge= True)
```

#### FACTOR PLOT

```python
sns.factorplot(x='day', y='total_bill', data= tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data= tips, kind='violin')
```

#### SWARM PLOT

```python
#draws a categorical scatterplot with non overlapping points
sns.swarmplot(x='day', y='total_bill', data= tips)
```
